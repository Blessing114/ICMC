```python
import numpy as np
import pandas as pd
import os
import networkx as nx
import re
import random
```

# 生成先增后减文件_add_minus

```python
file=os.listdir('./final_file')
# file=['PLG_1000_2776.txt']
for file_name in file:
    if file_name.endswith('.txt'):
        data=pd.read_csv('./final_file/'+file_name,sep='\t',header=None)
        a=['+']*len(data)
        a.extend(['-']*len(data))
        a=np.array(a)
        data_end=pd.DataFrame()
        data_end[0]=a
        in_=data[0].values
        out_=data[1].values
        data_end[1]=np.hstack((in_,in_[::-1]))
        data_end[2]=np.hstack((out_,out_[::-1]))
        name=file_name.strip('.txt')
        print(name)
        data_end.to_csv('./add_minus_file'+name+'_add_minus.txt',sep='\t',index=None,header=None)
```
